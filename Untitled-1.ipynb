```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import os
from tensorflow.keras.preprocessing.image import array_to_img

from LoadDataset import load_sewer_dataset, load_images, read_filepaths
import shutil
```

```python
data_dir = 'SewerImgs/MajorityNoText'
classes = ['Blurred', 'Sharp']

names, X_test, y_test = load_sewer_dataset(data_dir, classes)
```

```python
save_dir = 'Results'

model = 'AugmentedMaxResNet152'
model = 'Models/' + model + '.pkl'

with open(model, 'rb') as file:
    model = pickle.load(file)

y_pred_prob = model.predict(X_test)
y_pred = np.where(y_pred_prob > 0.5, 1, 0)

# True Positives: both y_true and y_pred are 1
TP_indices = [i for i, (yt, yp) in enumerate(zip(y_test, y_pred)) if yt == 1 and yp == 1]

# True Negatives: both y_true and y_pred are 0
TN_indices = [i for i, (yt, yp) in enumerate(zip(y_test, y_pred)) if yt == 0 and yp == 0]

# False Positives: y_true is 0 but y_pred is 1
FP_indices = [i for i, (yt, yp) in enumerate(zip(y_test, y_pred)) if yt == 0 and yp == 1]

# False Negatives: y_true is 1 but y_pred is 0
FN_indices = [i for i, (yt, yp) in enumerate(zip(y_test, y_pred)) if yt == 1 and yp == 0]
```

```python
for i, name in enumerate(names):
    if i in TP_indices:
        folder = 'TP'
    elif i in TN_indices:
        folder = 'TN'
    elif i in FP_indices:
        folder = 'FP'
    else: folder = 'FN'

    src_path = os.path.join('SewerImgs/NoText', name)
    save_path = os.path.join(save_dir, folder, name)

    shutil.copy2(src_path, save_path)
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def load_images2(df):

    # Create arrays to store images and image labels
    names = []
    images = []
    labels = []

    # Loop over all rows (images) in the dataframe to load them
    for index, row in df.iterrows():

        # Load the image
        img = load_img(row['file_path'])
        img_array = img_to_array(img)

        # Store the image and image label
        names.append(row['file_name'])
        images.append(img_array)
        labels.append(row['label'])

    # Return the images and image labels
    return np.array(names), images, np.array(labels)
```

```python
dir = 'Results'

folders = ['FN', 'FP', 'TN', 'TP']

df = read_filepaths(dir, folders)
df['Correctly Classified'] = [True if lab.startswith('T') else False for lab in df['label']]

names, imgs, labels = load_images2(df)
```

```python
imgs_int = [img.astype(np.int64) for img in imgs]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_mean_var_intensity(image):
    mean_intensities = [np.mean(image[:, :, c]) for c in range(image.shape[2])]
    var_intensities = [np.var(image[:, :, c]) for c in range(image.shape[2])]
    return mean_intensities + var_intensities

# Calculate the mean intensities for all images
mean_intensities = [calculate_mean_var_intensity(img) for img in imgs_int]

# Create a DataFrame
df_color = pd.DataFrame(mean_intensities, columns=['Mean_R', 'Mean_G', 'Mean_B', 'Var_R', 'Var_G', 'Var_B'])

merged_df = pd.merge(df, df_color, left_index = True, right_index = True)

sns.histplot(data = merged_df, x = 'Mean_R', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Mean Red Intensity on the Model')
plt.xlabel('Mean Red Intensity')
plt.savefig('MeanR.png')
plt.show()


sns.histplot(data = merged_df, x = 'Mean_G', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Mean Green Intensity on the Model')
plt.xlabel('Mean Green Intensity')
plt.savefig('MeanG.png')
plt.show()

sns.histplot(data = merged_df, x = 'Mean_B', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Mean Blue Intensity on the Model')
plt.xlabel('Mean Blue Intensity')
plt.savefig('MeanB.png')
plt.show()

sns.histplot(data = merged_df, x = 'Var_R', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Variance of Red Intensity on the Model')
plt.xlabel('Variance of Red Intensity')
plt.savefig('VarR.png')
plt.show()

sns.histplot(data = merged_df, x = 'Var_G', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Variance of Green Intensity on the Model')
plt.xlabel('Variance of Green Intensity')
plt.savefig('VarG.png')
plt.show()

sns.histplot(data = merged_df, x = 'Var_B', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Variance of Blue Intensity on the Model')
plt.xlabel('Variance of Blue Intensity')
plt.savefig('VarB.png')
plt.show()
```

```python
import tensorflow as tf
gray_imgs = [tf.image.rgb_to_grayscale(img) for img in imgs]
```

```python
average_brightness = [tf.reduce_mean(img).numpy() for img in gray_imgs]
brightness_std_dev = [tf.math.reduce_std(img).numpy() for img in gray_imgs]
```

```python
merged_df['average_brightness'] = average_brightness
merged_df['brightness_stdev'] = brightness_std_dev
```

```python
sns.histplot(data = merged_df, x = 'average_brightness', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Mean Brightness on the Model')
plt.xlabel('Mean Brightness')
plt.savefig('MeanBrightness.png')
plt.show()

sns.histplot(data = merged_df, x = 'brightness_stdev', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of St. Dev. of Brightness on the Model')
plt.xlabel('St. Dev. Brightness')
plt.savefig('StDevBrightness.png')
plt.show()
```

```python
from scipy.stats import entropy

gray_imgs = [tf.squeeze(img).numpy() for img in gray_imgs]
entropy_ = []

for img in gray_imgs:
    histogram, bin_edges = np.histogram(img, bins=256, range=(0, 1), density=True)
    entropy_.append(entropy(histogram))

merged_df['entropy'] = entropy_
```

```python
sns.histplot(data = merged_df, x = 'entropy', hue = 'Correctly Classified', stat = 'density', common_norm = False)
plt.title('Effect of Entropy on the Model')
plt.xlabel('Entropy')
plt.savefig('Entropy.png')
plt.show()
```
